# Files

## Writing

```python
f = open("message.txt", "w")

f.write("Hello World!")

f.close() # don't forget to close!
```

## Reading

```python
f = open("message.txt", "r")

m = f.read()

f.close()

print(m)
```

## Writing multiple lines

```python
from math import sqrt

f = open("square_roots.txt", "w")

for i in range(0,101):
    root = sqrt(i)
    f.write(str(root) + "\n")

f.close() # don't forget to close!

```

## Reading multiple lines

```python
f = open("square_roots.txt", "r")

roots_str = f.read().split("\n")

print(roots_str)
```

# Plotting data from a file

## Parsing a text file of data

```python
f = open("square_roots.txt", "r")

roots_str = f.read().split("\n")[:-1] # get rid of extra new line at the end

roots = [float(r) for r in roots_str] # cast all the strings into floats

print(roots)
```

## Plotting

```python
import matplotlib.pyplot as plt

plt.plot(roots)
```

## Setting axis limits

How do we look at the square roots for numbers > 10?


```python
plt.plot(roots)
plt.xlim([10,100])
```

## Making labels and titles

```python
plt.plot(roots)
plt.xlabel('x')
plt.ylabel('sqrt(x)')
plt.title('The Square Root Function')
```
